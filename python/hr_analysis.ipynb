```python
# Import Lib
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-v0_8')
sns.set_palette('Set2')
```

```python
# Load Data
overall=pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/HR Analytics/attrition_overall.csv')
```

```python
dept=pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/HR Analytics/attrition_by_department.csv')
```

```python
jobrole=pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/HR Analytics/attrition_by_jobrole.csv')
```

```python
gender=pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/HR Analytics/attrition_by_gender.csv')
```

```python
ageband=pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/HR Analytics/attrition_by_ageband.csv')
```

```python
joblevel = pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/HR Analytics/attrition_by_joblevel.csv')
```

```python
overtime=pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/HR Analytics/attrition_by_overtime.csv')
```

```python
promo=pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/HR Analytics/attrition_by_promotion_training.csv')
```

```python
travel=pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/HR Analytics/attrition_by_travel_distance.csv')
```

```python
# Attrition Overall
fig, ax=plt.subplots(figsize=(5,5))
labels=['Active','Attrition']
sizes = [overall['total_employee'][0] - overall['total_attrition'][0],
         overall['total_attrition'][0]]
ax.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90,colors=['#66c2a5','#fc8d62'])
ax.set_title('Overall Attrition Rate')
plt.savefig('attrition_overall.png',bbox_inches='tight')
plt.show()
```

```python
# Attrition by Department
plt.figure(figsize=(8,5))
sns.barplot(data=dept,x='Department',y='attrition_rate_pct',palette='Set2')
plt.title('Attrition Rate by Department')
plt.ylabel('Attration Rate (%)')
plt.xticks(rotation=45)
plt.savefig('attrition_by_department.png',bbox_inches='tight')
plt.show()
```

```python
# Attrition by Job Role
plt.figure(figsize=(10,6))
sns.barplot(data=jobrole,x='JobRole',y='attrition_rate_pct')
plt.title('Attrition by Job Role')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=45,ha='right')
plt.savefig('attrition_by_jobrole.png',bbox_inches='tight')
plt.show()
```

```python
# Attrition by Gender
plt.figure(figsize=(5,5))
sns.barplot(data=gender,x='Gender',y='attrition_rate_pct')
plt.title('Attrition Rate by Gender')
plt.ylabel('Attrition Rate (%)')
plt.savefig('attrition_by_gender.png',bbox_inches='tight')
plt.show()
```

```python
# Attrition by AgeBand
plt.figure(figsize=(7,5))
sns.barplot(data=ageband,x='AgeBand',y='attrition_rate_pct')
plt.title('Attrition by Age Band')
plt.ylabel('Attrition Rate (%)')
plt.savefig('attrition_by_ageband.png',bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.lineplot(data=joblevel, x="JobLevel", y="avg_attrition", marker="o")
plt.title("Average Attrition Rate by Job Level")
plt.ylabel("Avg Attrition Rate (%)")
plt.savefig('attrition_by_joblevel.png', bbox_inches="tight")
plt.show()
```

```python
#  Attrition by Overtime
plt.figure(figsize=(6,5))
sns.barplot(data=overtime, x='OverTimeFlag', y='attrition_rate_pct')
plt.title('Attrition by Overtime Status')
plt.ylabel('Attrition Rate (%)')
plt.xticks([0,1], ['No Overtime','Overtime'])
plt.savefig('attrition_by_overtime.png', bbox_inches='tight')
plt.show()
```

```python
#  Attrition by Promotion & Training
plt.figure(figsize=(8,5))
sns.scatterplot(data=promo, x='avg_years_since_promo', y='attrition_rate_pct',
                size='avg_training', hue='JobLevel', palette='Set2', sizes=(50,300))
plt.title('Attrition vs Promotion & Training')
plt.xlabel('Years Since Last Promotion')
plt.ylabel('Attrition Rate (%)')
plt.savefig('attrition_by_promotion_training.png', bbox_inches='tight')
plt.show()
```

```python
#  Attrition by Travel & Distance
plt.figure(figsize=(8,5))
sns.barplot(data=travel, x='BusinessTravel', y='attrition_rate_pct')
plt.title('Attrition by Business Travel')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=30)
plt.savefig('attrition_by_travel_distance.png', bbox_inches='tight')
plt.show()
```
